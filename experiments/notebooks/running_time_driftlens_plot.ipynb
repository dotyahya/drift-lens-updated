```python
def format_fn(tick_val, tick_pos):
    if tick_val > 0:
        x = tick_val/1000
        return f'{int(x)}K'
    else:
        return tick_val
    
def plot_driftlens_performance_only(data1, data2, filename):
    import matplotlib.pyplot as plt
    import numpy as np

    # Setup the figure and axes
    fig, axs = plt.subplots(1, 2, figsize=(20, 6))

    # Plot settings
    color = 'tab:blue'
    marker = 'o'
    linestyle = '-'

    # Extract and plot DriftLens data for the first dataset
    window_sizes_1 = sorted(data1.keys())
    means_1 = [data1[ws]['DriftLens']['mean'] for ws in window_sizes_1]
    stds_1 = [data1[ws]['DriftLens']['std'] for ws in window_sizes_1]
    #xlabels = ['{:,.2f}'.format(x) + 'K' for x in window_sizes_1/1000]
    axs[0].plot(window_sizes_1, means_1, linestyle, marker=marker, markersize=16, color=color, label='DriftLens', linewidth=3.5, alpha=0.8)
    axs[0].fill_between(window_sizes_1, np.subtract(means_1, stds_1), np.add(means_1, stds_1), color=color, alpha=0.2)

    axs[0].xaxis.set_major_formatter(format_fn)
    axs[0].set_ylim(top=0.5)
    axs[0].set_xlabel('Reference Window Size ($m_b$)', fontsize=28, labelpad=10)
    axs[0].set_ylabel('Mean Running Time (s)', fontsize=28)
    axs[0].set_title('a) Running Time for Reference Window Size', fontsize=30, pad=24)
    axs[0].grid(True)

    # Extract and plot DriftLens data for the second dataset
    window_sizes_2 = sorted(data2.keys())
    means_2 = [data2[ws]['DriftLens']['mean'] for ws in window_sizes_2]
    stds_2 = [data2[ws]['DriftLens']['std'] for ws in window_sizes_2]
    axs[1].plot(window_sizes_2, means_2, linestyle, marker=marker, markersize=16, color=color, label='DriftLens', linewidth=3.5, alpha=0.8)
    axs[1].fill_between(window_sizes_2, np.subtract(means_2, stds_2), np.add(means_2, stds_2), color=color, alpha=0.2)
    axs[1].xaxis.set_major_formatter(format_fn)
    axs[1].set_ylim(top=0.5)
    axs[1].set_xlabel('Datastream Window Size ($m_w$)', fontsize=28, labelpad=10)
    axs[1].set_title('b) Running Time for Datastream Window Size', fontsize=30, pad=24)
    axs[1].grid(True)

    # Adjusting plot aesthetics
    for ax in axs:
        ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey', alpha=0.7)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(True)
        ax.spines['bottom'].set_visible(True)
        ax.tick_params(axis='both', which='major', labelsize=26)

    # Setting the legend
    fig.legend(['DriftLens'], loc='upper center', bbox_to_anchor=(0.52, 0.1), fontsize=28)

    # Adjust layout to prevent overlap
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    # Save the figure
    plt.savefig(f'{filename}.pdf', format='pdf', bbox_inches='tight')
```

```python
plot_driftlens_performance_only(data1, data2, filename="running-time-driflens")
```
