```python
def plot_dual_drift_detector_performance_improved(data1, data2, data3, filename):
    import matplotlib.pyplot as plt
    import numpy as np

    # Prepare plot with two subplots side-by-side
    fig, axs = plt.subplots(1, 3, figsize=(20, 6))
    
    # Define your line styles
    custom_dash = (0, (3, 10, 1, 10))  # Custom dash pattern
    line_styles = ['--', '-', ':', '-', '-.']  # The last one is a custom dash pattern


    # Define markers and colors for better differentiation
    markers = ['d', 'o', 's', 'o', 'P']
    colors = ['tab:green', 'tab:blue', 'tab:purple', 'tab:orange', 'tab:red']
    techniques = list(set(list(data1[next(iter(data1))].keys()) + list(data2[next(iter(data2))].keys())))
    
     # Plot for the first dataset
    for idx, technique in enumerate(techniques):
        window_sizes = sorted(data1.keys())
        means = [data1[ws][technique]['mean'] for ws in window_sizes if technique in data1[ws]]
        stds = [data1[ws][technique]['std'] for ws in window_sizes if technique in data1[ws]]
        axs[0].plot(window_sizes, means, line_styles[idx % len(line_styles)], marker=markers[idx], markersize=10, color=colors[idx], label=technique, linewidth=2.5, alpha=0.8)
        axs[0].fill_between(window_sizes, np.subtract(means, stds), np.add(means, stds), color=colors[idx], alpha=0.2)

    axs[0].set_xlabel('Reference Window Size', fontsize=16, labelpad=10)
    axs[0].set_title('a) Running Time for Reference Window Size', fontsize=18, pad=20)
    axs[0].set_ylabel('Mean Running Time (s)', fontsize=16)
    axs[0].set_xticks(sorted(data1.keys()))
    axs[0].grid(True)

    

    # Plot for the first dataset
    for idx, technique in enumerate(techniques):
        #if technique != "MMD":
        window_sizes = sorted(data2.keys())
        means = [data2[ws][technique]['mean'] for ws in window_sizes if technique in data2[ws]]
        stds = [data2[ws][technique]['std'] for ws in window_sizes if technique in data2[ws]]
        axs[1].plot(window_sizes, means, line_styles[idx % len(line_styles)], marker=markers[idx], markersize=10, color=colors[idx], label=technique, linewidth=2.5, alpha=0.8)
        axs[1].fill_between(window_sizes, np.subtract(means, stds), np.add(means, stds), color=colors[idx], alpha=0.2)
    
    axs[1].set_xlabel('Datastream Window Size', fontsize=16, labelpad=10)
    axs[1].set_title('b) Running Time for Datastream Window Size', fontsize=18, pad=20)
    axs[1].set_xticks(sorted(data2.keys()))
    axs[1].grid(True)

    
    # Show y-ticks on the right plot
    axs[1].tick_params(labelleft=True)

    
    # Plot for the third dataset
    for idx, technique in enumerate(techniques):
        window_sizes = sorted(data3.keys())
        means = [data3[ws][technique]['mean'] for ws in window_sizes if technique in data3[ws]]
        stds = [data3[ws][technique]['std'] for ws in window_sizes if technique in data3[ws]]
        axs[2].plot(window_sizes, means, line_styles[idx % len(line_styles)], marker=markers[idx], markersize=10, color=colors[idx], label=technique, linewidth=2.5, alpha=0.8)
        axs[2].fill_between(window_sizes, np.subtract(means, stds), np.add(means, stds), color=colors[idx], alpha=0.2)

    axs[2].set_xlabel('Embedding dimensionality', fontsize=16, labelpad=10)
    axs[2].set_title('c) Running Time for Embdding Dimensionality', fontsize=18, pad=20)
    axs[2].set_xticks(sorted(data3.keys()))
    axs[2].grid(True)

    # Show y-ticks on the right plot
    axs[2].tick_params(labelleft=True)
    
    # Set grid and spines
    for ax in axs:
        ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey', alpha=0.7)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(True)
        ax.spines['left'].set_visible(True)
        ax.tick_params(axis='both', which='major', labelsize=14)

    # Configure legend
    handles, labels = axs[0].get_legend_handles_labels()
    #fig.legend(handles, labels, loc='upper center', ncol=len(techniques), bbox_to_anchor=(0.5, 1.05), fontsize=16)

    #plt.tight_layout(rect=[0, 0, 1, 0.95])
    
    fig.legend(handles, labels, loc='upper center', ncol=len(techniques), bbox_to_anchor=(0.5, -0.01), fontsize=16)

    # Adjust tight_layout to make space for the legend at the bottom
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    
    #plt.show()
    plt.savefig(f'{filename}.pdf', format='pdf', bbox_inches='tight')

```
